```python
import sys
import pandas as pd
import numpy as np
from pandas import Series
data_orig = pd.read_csv('train.tsv',sep = '\t')
data_test_orig = pd.read_csv('validation.tsv',sep = '\t')
data = pd.read_csv('train.tsv',sep = '\t')
data_test = pd.read_csv('validation.tsv',sep = '\t')
```

```python
import re

#TRAIN
data['tags'] = data['tags'].str.replace(r"[,\'\[\]]",'')
data['title'] = data['title'].str.replace(r"[\']",'')
data['title'] = data['title'].str.lower()
data['title'] = data['title'].str.replace(r"[^a-z0-9 #+]",' ')
data['title'] = data['title'].str.replace('\s+',' ')

#TEST
data_test['tags'] = data_test['tags'].str.replace(r"[,\'\[\]]",'')
data_test['title'] = data_test['title'].str.replace(r"[\']",'')
data_test['title'] = data_test['title'].str.lower()
data_test['title'] = data_test['title'].str.replace(r"[^a-z0-9 #+]",' ')
data_test['title'] = data_test['title'].str.replace('\s+',' ')
```

```python
import nltk
stopword = nltk.corpus.stopwords.words('english')
wn = nltk.WordNetLemmatizer()
token = nltk.tokenize.WhitespaceTokenizer()

def token_stopword_lemm(title):        
    token_list = token.tokenize(title)
    not_stopword_list = [word for word in token_list if word not in stopword]
    title = " ".join([wn.lemmatize(word) for word in not_stopword_list])
    return title

#TRAIN
data['title'] = data['title'].apply(lambda x: token_stopword_lemm(x))
data.head(10)
```

```python
#TEST
data_test['title'] = data_test['title'].apply(lambda x: token_stopword_lemm(x))
data_test.head(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(min_df = 2, max_df = 0.5,ngram_range=(1,2))
feat_extract = tfidf.fit(data['title'])

#TRAIN
x_train = tfidf.transform(data['title'])
print(x_train.shape)
print(tfidf.get_feature_names())
```

```python
#TEST
x_test = tfidf.transform(data_test['title'])
print(x_test.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
label_c_train = CountVectorizer(token_pattern = "[\S]+")
feat_extract = label_c_train.fit(data['tags'])

#TRAIN
x_label = label_c_train.transform(data['tags'])
print(x_label.shape)
print(x_label)
print(label_c_train.get_feature_names())
```

```python
#TEST
x_test_label = label_c_train.transform(data_test['tags'])
print(x_test_label.shape)
print(x_test_label)
```

```python
from scipy import sparse


#TRAIN
label_train = x_label.todense()

#TEST
label_test = x_test_label.todense()
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

#Train
#model_rand_forest = RandomForestClassifier()
# parameters = {'n_estimators' : [50,100], 'max_depth'  : [30,80,None] }

# model = GridSearchCV(model_rand_forest,parameters,n_jobs = -1)
# model.fit(x_train,label_train)
```

```python
#TRAIN VALIDATION SPLIT

feat_train,x_validation,feat_label,label_validation = train_test_split(x_train,label_train,test_size = 0.25)
max_acc = 0
for c in [0.1,1,10,100]:
    for p in ['l1','l2']:
        Lregression = LogisticRegression(multi_class = 'ovr',C=c,penalty = p)
        OneVRest = OneVsRestClassifier(Lregression, n_jobs=-1)
        model = OneVRest.fit(feat_train,feat_label)
        label_validation_predict = model.predict(x_validation)
        accuracy = accuracy_score(label_validation,label_validation_predict)
        print('C := ',c, '  Penalty:= ',p,'   Accuracy= ',accuracy)
        if(accuracy>=max_acc):
            max_acc = accuracy
            optimal_C = c
            optimal_P = p
        
Lregression = LogisticRegression(multi_class = 'ovr',C = optimal_C,penalty = optimal_P)
OneVRest = OneVsRestClassifier(Lregression, n_jobs=-1)
model = OneVRest.fit(feat_train,feat_label)
label_validation_predict = model.predict(x_validation)
accuracy = accuracy_score(label_validation,label_validation_predict)
print('Accuracy= ',accuracy) 

# print(optimal_n_est)
# print()
# print(optimal_d)
# print()
# print(max_acc)
        
        
```

```python
#TRAIN PREDICTION
label_train_predict = model.predict(x_train)
label_train_predict_inversed = label_c_train.inverse_transform(label_train_predict)
label_train_inversed = label_c_train.inverse_transform(label_train)
for i in range(30):
    print('Title:\t{}\nTrue labels:\t{}\nPredicted labels:\t{}\n\n'.format(
        data['title'].iloc[i],
        ','.join(label_train_inversed[i]),
        ','.join(label_train_predict_inversed[i])
    ))
```

```python
#TEST PREDICTION
label_test_predict = model.predict(x_test)
label_test_predict_inversed = label_c_train.inverse_transform(label_test_predict)
label_test_inversed = label_c_train.inverse_transform(label_test)
for i in range(30):
    print('Title:\t{}\nTrue labels:\t{}\nPredicted labels:\t{}\n\n'.format(
        data_test['title'].iloc[i],
        ','.join(label_test_inversed[i]),
        ','.join(label_test_predict_inversed[i])
    ))
```
